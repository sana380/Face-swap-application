```python
# Set style for plots
plt.style.use('default')
sns.set_palette("husl")

# Key vars for analysis
key_vars = ['tempo_como_cliente_meses', 'n_compras_ano', 'faturacao_anual_eur',
            'lucro_anual_eur', 'divida_ativa_eur', 'dias_atraso', 'limite_credito_eur',
            'avaliacao_cliente_1a5', 'custo_suporte_eur']
```

```python
# 2. Correlation heatmap (focus on debt/delays)
numeric_df = df[key_vars]
corr = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix (Key Financial Vars)')
plt.savefig('eda_correlations.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 3. Boxplots: Debt and Delays by Sector
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(data=df, x='setor', y='divida_ativa_eur', ax=ax1)
ax1.set_title('Active Debt by Sector')
ax1.tick_params(axis='x', rotation=45)

sns.boxplot(data=df, x='setor', y='dias_atraso', ax=ax2)
ax2.set_title('Payment Delays by Sector')
ax2.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig('eda_by_sector.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 4. Sector summary table
sector_summary = df.groupby('setor')[['faturacao_anual_eur', 'lucro_anual_eur', 'divida_ativa_eur', 'dias_atraso', 'avaliacao_cliente_1a5']].mean().round(2)
print(sector_summary)
sector_summary.to_csv('sector_summary.csv')
```

```python
# Prepare data for clustering
features = ['tempo_como_cliente_meses', 'n_compras_ano', 'faturacao_anual_eur',
            'lucro_anual_eur', 'divida_ativa_eur', 'dias_atraso', 'limite_credito_eur',
            'avaliacao_cliente_1a5', 'custo_suporte_eur']
X = df[features]
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Elbow method to find optimal k
inertia = []
sil_scores = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
# Plot elbow
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(k_range, inertia, marker='o')
ax1.set_title('Elbow Method')
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia')

ax2.plot(k_range, sil_scores, marker='o')
ax2.set_title('Silhouette Scores')
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
plt.tight_layout()
plt.savefig('clustering_elbow.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Fit K=3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['segment'] = kmeans.fit_predict(X_scaled)
```

```python
# Segment profiles (means)
segment_profile = df.groupby('segment')[features].agg(['mean', 'count']).round(2)
print(segment_profile)
```

```python
# Visualize clusters (PCA for 2D plot)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['segment'], palette='Set1')
plt.title('Client Segments (PCA Projection)')
plt.savefig('clustering_pca.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Sector distribution per segment
segment_sector = pd.crosstab(df['segment'], df['setor'], normalize='index').round(3) * 100
print(segment_sector)
segment_profile.to_csv('segment_profiles.csv')
```
